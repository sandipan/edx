# Registering for the Course Autograder

This notebook registers you for the course autograder. You need to use the autograder to get a grade for each lab.

**You will only need to use this notebook once.**

This notebook will help you create an _autograder token_. You will use that token when you submit each lab for grading, but you'll submit each lab using the
[autograder notebook](https://raw.githubusercontent.com/spark-mooc/mooc-setup/master/cs120_autograder_simple.dbc).

If you're interested in more details on the autograder, see the [Complete Autograder notebook](https://raw.githubusercontent.com/spark-mooc/mooc-setup/master/cs120_autograder_complete.dbc).

## Step 2: Import the Autograder Notebook

Import a copy of the autograder notebook:
